```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import re
```

```python
def crawl(alist, seed, max, step):
  #open and read the page
  openurl = urlopen(seed)
  obj = soup(openurl.read())

  #get all links in current page
  templist = []
  for link in obj.findAll('article',{'class':'ph_newsfeed_d article_inview list-content__item'}):
    news = link.find('a')
    if 'href' in news.attrs:
      #only add link that hasn't been in the list
      if news not in alist:
        templist.append(news)

  alist.extend(templist)
  #step to the next url in the list if not already max

  print("from " + seed + " getting " + str(len(templist))+ " ")

  if step < max:
    for alink in templist:
        aurl = templist.pop(0).attrs['href']
        print("going to: " + aurl)
        alist.extend(crawl(alist, aurl, max, step+1))
  return alist
```

```python
queue = []
seed = "https://news.detik.com/berita"
max = 1
step = 0
datas = crawl(queue, seed, max, step)
```
